```python
# ============================================
# SCHEMA RESOLVER UTILITY
# ============================================

def resolve_schema(
    root_element,
    registry_table
):

    registry_df = spark.table(
        registry_table
    )

    schema_df = registry_df.filter(
        registry_df.root_element ==
        root_element
    )

    if schema_df.count() == 0:

        raise Exception(
            f"No schema found for root element: "
            f"{root_element}"
        )

    return schema_df
```
