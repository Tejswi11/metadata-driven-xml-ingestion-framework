```python

# TYPE CASTING ENGINE


from pyspark.sql.functions import col


def apply_type_casting(
    df,
    target_table
):

    target_schema = spark.table(
        target_table
    ).schema

    for field in target_schema:

        column_name = field.name

        data_type = field.dataType

        if column_name in df.columns:

            df = df.withColumn(
                column_name,
                col(column_name).cast(
                    data_type
                )
            )

    return df
```

```python
# ============================================
# DELTA LOADER UTILITY
# ============================================

def load_to_delta(
    df,
    target_table,
    load_mode="append"
):

    df.write \
        .format("delta") \
        .mode(load_mode) \
        .saveAsTable(
            target_table
        )

    print(
        f"Data loaded into: {target_table}"
    )
```
