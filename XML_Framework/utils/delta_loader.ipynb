```python
# ============================================
# DELTA LOADER UTILITY
# ============================================

def load_to_delta(
    df,
    target_table,
    load_mode="append"
):

    df.write \
        .format("delta") \
        .mode(load_mode) \
        .saveAsTable(
            target_table
        )

    print(
        f"Data loaded into: {target_table}"
    )
```
