```python
# ============================================
# FILE ARCHIVER
# ============================================

def archive_file(
    source_file,
    archive_volume
):

    file_name = source_file.split("/")[-1]

    target_file = (
        f"{archive_volume}/{file_name}"
    )

    dbutils.fs.mv(
        source_file,
        target_file
    )

    print(
        f"Archived: {file_name}"
    )
```
