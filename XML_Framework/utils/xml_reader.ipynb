```python

# ============================================
# XML READER UTILITY
# ============================================

import os


def discover_xml_files(xml_volume_path):

    xml_files = []

    files = dbutils.fs.ls(xml_volume_path)

    for file in files:

        if file.path.lower().endswith(".xml"):

            file_name = file.name

            file_path = file.path

            table_name = os.path.splitext(
            file_name
            )[0]

            xml_files.append({

                "file_name": file_name,

                "file_path": file_path,

                "table_name": table_name

            })

    return xml_files
```
