```python
# xml structure explore 
def explore_xml_structure (
    root,
    parent = None,
    level=0,
    path=""
):
    structure = []

    current_path = (
        root.tag
        if path == ""
        else f"{path}.{root.tag}"
    )

    structure.append({
        "element_name": root.tag,
        "parent": parent,
        "level": level,
        "path": current_path
    })

    for child in root:
        structure.extend(
            explore_xml_structure(
                root = child,
                parent = root.tag,
                level = level + 1,
                path = current_path
            )
        )
    return structure
```
